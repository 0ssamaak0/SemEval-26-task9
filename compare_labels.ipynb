```python
import pandas as pd
```

```python
labels = pd.read_csv('./dev_phase/subtask1/dev/arb.csv')
bert_based_uncased = pd.read_csv('./results/ar_bert_based_uncased.csv')
MARBERTv2 = pd.read_csv('./results/ar_MARBERTv2.csv')
```

```python
# merge all of them based on id
bert_based_uncased = bert_based_uncased.rename(columns={"polarization": "bert_base_label"})
MARBERTv2 = MARBERTv2.rename(columns={"polarization": "marbert_label"})
merged_df = pd.merge(labels, bert_based_uncased, on='id', how='left')
merged_df = pd.merge(merged_df, MARBERTv2, on='id', how='left')
# Save it into combined labels .csv
merged_df.to_csv('./results/combined_labels.csv', index=False)
```
